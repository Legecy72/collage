```python
import math

def sigmoid(x): return 1 / (1 + math.exp(-x))
def tanh(x): return math.tanh(x)
```

```python
Wf, Wi, Wg, Wo = 0.5, 0.6, 0.4, 0.7
Uf, Ui, Ug, Uo = 0.3, 0.4, 0.2, 0.5
bf, bi, bg, bo = 0.1, 0.1, 0.0, 0.1
Wy, by = 1.5, 0.3

h, c = 0.0, 0.0
inputs = [1, 2, 3, 4]
```

```python
for t, x in enumerate(inputs, 1):
    f = sigmoid(Wf*x + Uf*h + bf)
    i = sigmoid(Wi*x + Ui*h + bi)
    g = tanh(Wg*x + Ug*h + bg)
    c = f*c + i*g
    o = sigmoid(Wo*x + Uo*h + bo)
    h = o * tanh(c)
    print(f"t={t}: f={f:.4f}, i={i:.4f}, g={g:.4f}, c={c:.4f}, o={o:.4f}, h={h:.4f}")
```

```python
y_hat = Wy * h + by
print(f"y_hat = {Wy} * {h:.4f} + {by} = {y_hat:.4f}")
```
